```
import pandas as pd
pd.set_option('display.max_colwidth',None)
```

```
main_result= pd.DataFrame(columns=['file_name','height','width','annotations'])
main_result['annotations']=main_result['annotations'].astype('object')
```

```
cat_dict = {"pedestrian":"0","car":"1","van":"2","bus":"3","truck":"4","motor":"5","bicycle":"6","tricycle":"7","other":"8"}

import cv2
from google.colab.patches import cv2_imshow
```

```
header_list = ["xmin","ymin","xmax","ymax","category_id"]
```

```
import os
file_list = os.listdir('/content/Json/testTxt')
for i in file_list:
  print(i)
```

```
name_list = []
for i in file_list:
  name_list.append(i.strip('.txt'))
for i in name_list:
  print(i)
```

```
path_list = []
path = '/content/Json/testTxt/'
for i in file_list:
  path_list.append(os.path.join(path,i))
```

```
for i,j in zip(path_list,name_list):
  df = pd.read_csv(i,header=None,index_col=False,names=header_list)
  df['height'] = abs(df['ymax']-df['ymin'])
  df['width'] = abs(df['xmax']-df['xmin'])
  df = df[['xmin','ymin','width','height','category_id']]
  df1=df

  df1['bbox'] = df1.iloc[:,0:4].values.tolist()
  df1['bbox_mode'] = 1
  df1 = df1.replace({'category_id':cat_dict})
  df1=df1[['bbox','bbox_mode','category_id']]
  annotations = df1.T.to_dict().values()
  l=[]
  for k in annotations:
    l.append(k)
  res = pd.DataFrame(columns=['file_name','height','width','annotations'])
  res['annotations'] = res['annotations'].astype('object')
  res.at[0,'file_name'] = i[:-4]+'.jpg'
  res.at[0,'annotations'] = l
  h = cv2.imread('/content/Json/Test/'+str(j)+'.jpg').shape[:2]
  res.at[0,'height'] = h[0]
  res.at[0,'width'] = h[1]

  main_result = main_result.append(res)
  main_result.reset_index(drop=True,inplace=True)
```

```
main_result.reset_index(inplace=True)
main_result.rename(columns={'index':"image_id"},inplace=True)
main_result.to_json("test.json",orient="records")
```
